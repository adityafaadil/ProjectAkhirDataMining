## Scraping Data

**Apabila dalam proses scraping error maka jalankan ulang cell no2 diatas .**

```python
# Buat objek Twint
c = twint.Config()

#Mencari tweet yang mengandung kata Gempa Bumi Jawa Barat, di, magnitude
c.Search = "Piala Dunia 2022"

#Jumlah Data
c.Limit = 10000

#Menyimpan data ke dalam file csv
c.Store_csv = True
c.Output = 'Data_Twitter.csv'

#Menjalankan scrapping
twint.run.Search(c)
```
